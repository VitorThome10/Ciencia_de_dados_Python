```python
# Calculando rentabilidade de um modelo hipotético mês a mês

import pandas as pd
from pandas_datareader import data as pdr
import numpy as np
import yfinance as yf
# Calculando o retorno acumulado dos dados

acoes01 = ["SANB11.SA","WEGE3.SA","ABEV3.SA", "PRIO3.SA"]

dados_acoes = yf.download(acoes01,"2019-01-01","2023-06-14")["Adj Close"]

retornos = dados_acoes.pct_change().dropna()

retornos

retornos["rent_modelo"] = retornos["WEGE3.SA"] * 0.25 + retornos["PRIO3.SA"]*0.25 + retornos["SANB11.SA"]*0.25 + retornos["SANB11.SA"]*0.25

retornos ["ano"] = retornos.index.year
retornos ["mes"] = retornos.index.month

retornos["rent_modelo"] = 1 + retornos["rent_modelo"]

retornos ["rent_MID"] = retornos.groupby(["ano", "mes"])["rent_modelo"].cumprod() - 1

retornos.groupby(["ano","mes"]).tail(1)[["ano","mes","rent_MID"]]
```
