```python
import pandas as pd
from pandas_datareader import data as pdr
import numpy as np
import yfinance as yf
```

```python
# Correlação móvel entre o SP500 e o fundo de comoditties

acoes = ["^BVSP", "MATB11.SA"]

dados_indices = yf.download(acoes, "2010-01-01" ,  "2022-08-31")["Adj Close"]

retornos = dados_indices.pct_change().dropna()

corr_movel = retornos["^BVSP"].rolling(252).corr(retornos["MATB11.SA"]).dropna()

corr_movel.plot()
```
