```python
## Criando o Dataframe para fazer a correlação
import matplotlib.pyplot as plt
import pandas as pd
dados_lucro3tentos = [1043, 701, 500, 314, 297]
preco_dolar = [5.30, 5.41, 5.48, 4.18, 3.79]
data = [2022, 2021, 2020, 2019, 2018]

df = pd.DataFrame({'coluna1': dados_lucro3tentos, 'coluna2':preco_dolar}, index=data )
df
```

```python


correlacao = df['coluna1'].corr(df['coluna2'])
print(f"A Correlação para a empresa 3tentos entre 2018 e 2022 entre o dolár e lucro é de {correlacao}")

```

```python
## Plotando em gráfico de barra
fig, ax = plt.subplots()

ax.bar(df.index, df["coluna1"], label="Coluna 1")
ax.bar(df.index, df["coluna2"], label="Coluna 2")  # Use alpha para transparência

```
