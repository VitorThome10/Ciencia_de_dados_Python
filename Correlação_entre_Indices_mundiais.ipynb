```python
import pandas as pd
from pandas_datareader import data as pdr
import numpy as np
import yfinance as yf
```

```python
# Calculando a correlacao entre indices e o IBOV durante 10 anos

acoes = ["^SET.BK","^KS11","^N225",
         "TA35.TA","^HSI", "^FTSE", "^BVSP", "^GSPC","^DJI","^JKSE","^TASI.SR"]

dados_acoes = yf.download(acoes, "2011-08-31","2022-08-31")["Adj Close"]

retornos = dados_acoes.pct_change().dropna() # Calculando os retornos das ações

retorno_ibov = retornos["^BVSP"]

retornos = retornos.drop("^BVSP", axis = 1)

corr_movel = retornos.rolling(252).corr(retorno_ibov)

corr_movel = corr_movel.dropna()

corr_movel
```

```python
corr_movel.plot()
```
