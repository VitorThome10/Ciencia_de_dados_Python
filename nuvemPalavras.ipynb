```python
from pypdf import PdfReader
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import matplotlib.pyplot as plt
```

```python
reader = PdfReader("/content/Release%20de%20Resultados%202T23.pdf")
text = ""
for page in reader.pages:
    text += page.extract_text()
```

```python
conectivos = set(STOPWORDS)
conectivos.update(["da", "meu", "em", "e", "você", "de", "ao", "os", 'na', 'o', '4T21', 'para', 'milhões',
                  'resultado', 'que', 'nas', 'dos', "n", "p", "4T22", "foi", 'ano', '3T22', 'resultados',
                  'trimestre', 'AH', 'nos', 'segmento', 'release', 'das', 'um', 'uma', '2T23', '2T22', 'mais'])

```

```python
nuvem_palavras = WordCloud(stopwords = conectivos, background_color = 'black',
                           width = 1920, height = 1080).generate(text)
```

```python
fig, ax = plt.subplots()

ax.imshow(nuvem_palavras)
ax.set_axis_off()
```
